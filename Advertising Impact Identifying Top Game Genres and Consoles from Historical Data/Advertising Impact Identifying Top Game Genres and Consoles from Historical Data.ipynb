#### Kolom rating

Pertama kita hitung jumlah data yang hilang di kolom ini.

```python
df_fixed['rating'].isna().sum()/len(df_fixed)
```

Terdapat 40% data, tidak memiliki nilai rating. Cukup besar, sehingga harus kita atasi.

```python
df_fixed['rating'].value_counts()
```

Berdasarkan informasi dari www.g2a.com tentang jenis rating ESRB:
1. E = Everyone (untuk semua usia)
2. E10 = Everyone over ten years old (Untuk usia di atas 10 tahun)
3. T = Teen (Untuk usia di atas 13 tahun)
4. M = Mature (Untuk usia di atas 17 tahun)
5. AO = Adult Only (Untuk usia di atas 18 tahun)
6. RP = Rating Pending (Masih menunggu penilaian)

Rating  k-a : kids to adult. Kids bisa kita anggap sebagai kategori "children" dengan rentang usia 6-13 tahun. Sehingga rating k-a akan dimasukkan sebagai kategori E (Everyone). Untuk saat ini ESRB tidka memiliki rating ini.

Rating EC : Early Childhood juga saat ini dimasukan kedalam rating E (Everyone).

```python
df_fixed.loc[df_fixed['rating'].isin(['EC','K-A']),'rating'] = 'E'
```

Sekarang kita akan melihat apakah ada rating yang kosong pada salah satu platform, tidak kosong di platform lain. Harusnya game yang sama meski beda platform ratingnya akan sama.

```python
no_rating_list = df_fixed.loc[df_fixed['rating'].isna(),'name']
```

```python
df_fixed.loc[df_fixed['name'].isin(no_rating_list) & ~df_fixed['rating'].isna(),:].drop_duplicates(subset='name')
```

Terdapat 359 data yang dapat kita isi rating nya.

```python
def getRating(data):
    name   = data['name']
    rating = df_fixed.loc[(df_fixed['name']==name) & (~df_fixed['rating'].isna()),:].drop_duplicates(subset='name')['rating']
    if len(rating)==1:
        return rating.iloc[0]
    else: return np.nan
```

```python
df_fixed.loc[df_fixed['rating'].isna(),'rating'] = df_fixed.loc[df_fixed['rating'].isna(),:].apply(getRating, axis = 1)
```

```python
df_fixed['rating'].isna().sum()/len(df_fixed)
```

Ternyata hanya berkurang 2.2% lumayan. Strategi selanjutnya dengna melihat rating berdasarkan genre game. Biasanya genre game berkolerasi tinggi terhadap rating game. Mari kita lihat.

```python
df_fixed.loc[df_fixed['rating']!='nan',['rating','genre','total_sales']].pivot_table(index='genre',columns='rating',values='total_sales',aggfunc='count')
```

Ternyata beberapa genre memang didominasi oleh salah-satu rating saja, meski memang beberapa genre memiliki lebih dari satu rating yang dominan. Saya akan menggunakan rating paling banyak dari setiap genre untuk mengisi nilai yang kosong pada kolom rating. Mari kita check modusnya rating dari masing-masing genre

```python
pvt_rating = df_fixed.pivot_table(index='genre',values='rating',aggfunc=lambda x: x.mode())
pvt_rating
```

```python
def getRatingbyMode(data):
    genre=data['genre']
    return pvt_rating.loc[genre,'rating']
```

```python
df_fixed.loc[df_fixed['rating'].isna(),'rating'] = df_fixed.loc[df_fixed['rating'].isna(),:].apply(getRatingbyMode, axis = 1)
```

```python
df_fixed.loc[df_fixed['rating'].isna(),'rating']
```

Nilai hilang pada kolom rating telah diisi oleh nilai modus berdasarkan Genre nya.

Dengan melihat ke-5 konsol sebelumnya saja sudah terdapat banyak sekali outlier, apalagi data keseluruhan konsol?. Sekarang tugas kita yaitu mengatasi outlier yang ada.

```python
def getIQR(console):
    q25 = np.quantile(df_fixed.loc[df_fixed['platform']==console,'total_sales'],.25)
    q75 = np.quantile(df_fixed.loc[df_fixed['platform']==console,'total_sales'],.75)
    iqr = (q75-q25)
    return q75, q25,iqr
def getLower(data):
    console = data['platform']
    UL, LL, iqr = getIQR(console)
    return LL - 1.5*iqr
def getUpper(data):
    console = data['platform']
    UL, LL, iqr = getIQR(console)
    return UL + 1.5*iqr
```

Menghitung batas atas dan bawah outlier dari masing-masing console

```python
batas_outlier = df_fixed.loc[:,['platform']]
batas_outlier.drop_duplicates(subset='platform',inplace=True)
batas_outlier.reset_index(drop=True, inplace=True)
batas_outlier.head()
```

```python
batas_outlier['batas_bawah'] = batas_outlier.apply(getLower, axis = 1)
batas_outlier['batas_atas'] = batas_outlier.apply(getUpper, axis = 1)
batas_outlier.head()
```

```python
def removeOutlier(data):
    platform = data['platform']
    sales    = data['total_sales']
    LL = batas_outlier.loc[batas_outlier['platform']==platform,'batas_bawah']
    UL = batas_outlier.loc[batas_outlier['platform']==platform,'batas_atas']
    UL = list(UL)[0]
    LL = list(LL)[0]
    if (sales > UL) or (sales < LL):
        return np.nan
    else:
        return sales
```

```python
df_no_outlier = df_fixed.copy()
df_no_outlier['total_sales'] = df_no_outlier.apply(removeOutlier, axis = 1)
```

```python
df_no_outlier.dropna(subset=['total_sales'],inplace=True)
df_no_outlier.loc[df_no_outlier['total_sales']>0,:]
```

```python
plt.figure(figsize=(20,20))
plt.title('Perbandingan Boxplot Seluruh Platform (Tanpa Outlier)',fontsize = 16
        )
plt.ylabel('Total Pendapatan Tiap Game, juta USD')
plt.xlabel('Platform')
sb.boxplot(data=df_no_outlier, x='platform', y='total_sales',
           order=df_no_outlier.groupby('platform')['total_sales'].median().
           reset_index(drop=False).
           sort_values(by='total_sales',ascending=False).platform)
plt.show()
```

### Rata-rata Rating Pengguna Xbox One dan PC

Hipotses rating pengguna adalah hipotesis dimana rating pengguna Xbox One dan PC sama.

- H0 : Rating pengguna Xbox One dan PC sama
- H1 : Rating pengguna Xbox One dan PC berbeda

```python
xbox = df_potensi_no.query("platform=='XOne'")['user_score'].dropna()
pc = df_potensi_no.query("platform=='PC'")['user_score'].dropna()
```

Pertama kita uji terlebih dahulu varians-nya baik melihat langsung secara value maupun menggunakan bantuan scipy.stats.levene.

```python
print(f'varians rating xbox {xbox.std()}')
print(f'varians rating pc {pc.std()}')
1-xbox.std()/pc.std()
```

Secara value perbedaanya cukup besar yaitu mencapai 16% mari kita uji dengan levene dengan toleransi 5%.

```python
stat, p = scipy.stats.levene(xbox, pc)
print(f'nilai p  = {p}')
if p < 0.05:
    print("Varians Xbox One dengan PC berbeda")
else:
    print("Varians Xbox One dengan PC sama") 
```
